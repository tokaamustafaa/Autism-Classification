# deep learning model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.optimizers import Adam , SGD
from tensorflow.keras.callbacks import EarlyStopping , ModelCheckpoint
from tensorflow.keras.regularizers import L1,L2
from kerastuner.tuners import RandomSearch
```

```python
x_train.shape[1]
```

```python
def build_model(hp):
    model=Sequential([
        Dense(units=hp.Int('units' , min_value=32 , max_value=64 , step=16) , activation=hp.Choice('activation' , ['relu' , 'sigmoid']) , input_shape=(x_train.shape[1],)),
        Dropout(rate=hp.Choice('rate' , [0.2 , 0.3 , 0.4 ,0.5])) ,
        Dense(units=hp.Int('units' , min_value=64 , max_value=128 , step=16) , activation=hp.Choice('activation' , ['tanh' , 'sigmoid'] )),
        Dropout(rate=hp.Choice('rate' , [0.2 , 0.3 , 0.4 ,0.5])) ,
        Dense(units=hp.Int('units' , min_value=128 , max_value=256 , step=64) , activation=hp.Choice('activation' , ['tanh' , 'sigmoid'] )),
        Dense(units=1 , activation='sigmoid')

    ])
    model.compile(optimizer=Adam( learning_rate=hp.Choice('learning_rate', values=[0.1, 0.2, 0.3])) ,  
              loss='binary_crossentropy' , 
              metrics=['accuracy']         
              )
    return model
```

```python
tuner=RandomSearch(
    build_model,
    objective='val_loss' , max_trials=10 , directory='lab5' , project_name='session_6'
    
)
early_stopping=EarlyStopping(patience=5 , monitor='val_loss' , restore_best_weights=True)
```

```python
checkpoints=ModelCheckpoint('best_model.keras' , monitor='val_loss' , save_best_only=True)
tuner.search(x_train , y_train , epochs=5 , validation_split=0.1 , callbacks=[early_stopping , checkpoints])
```
